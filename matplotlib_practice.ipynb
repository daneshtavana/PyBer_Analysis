```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

Annotate Charts Using the MATLAB Method

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston Area')
# plt.xlim("Jan", "June")
# plt.xlim(0, 5)
plt.ylim(5, 50)
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title("PyBare Fare by Month")
plt.legend()
plt.grid()
# plt.figure(figsize = (20,10))
# plt.subplots(figsize = (20,10))
# plt.show() NOTE: when I insert this the figure isn't printed?
plt.savefig("./Figure_1_Line")
```

Annotate Charts Using the Object-Oriented Interface

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots(figsize=(8, 4))
# Or replace above line with
# fig = plt.figure()
# ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston Area')
# ax.set_xlim(0, 5)
ax.set_ylim(5, 50)
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title("PyBare Fare by Month")
ax.legend()
ax.grid()
plt.savefig("./Figure_2_Line")                     
```

Create Bar Charts Using the MATLAB Approach

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston Area')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.savefig("./Figure_3_HBar")    
```

Create Bar Charts Using the Object-Oriented Approach

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='City of Chicago')
ax.invert_yaxis()
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_title("PyBare Fare by Month")
ax.legend()
plt.savefig("./Figure_4_HBar")                    #
```

5.1.7 Create Scatter Plots and Bubble Charts

```python
plt.plot(y_axis, x_axis, 'o', color="red", label='City of Chicago')
plt.gca().invert_yaxis()
# This also works
# plt.scatter(x_axis, y_axis, color="red", label='City of Chicago')# Create the plot.
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.xlim(0, 50)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.savefig("./Figure_5_Scatter")                    #
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

Create a Scatter Plot Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           color = "skyblue", edgecolor = "Black" , linewidths = 2,
           s = [i * 5 for i in y_axis], 
           label="Boston")
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_title("PyBare Fare by Month")
ax.legend()
# ax.grid()
plt.savefig("./Figure_6_Scatter")     
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
